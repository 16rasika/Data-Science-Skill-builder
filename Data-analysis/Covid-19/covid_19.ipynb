1.Top 10 covid confirmed country with there WHO region

```python
# Top 10 covid confirmed country
top_confirmed_cases = covid_19.sort_values('Confirmed',ascending = False)
# Selecting top ten rows and columns 
top_confirmed_cases = top_confirmed_cases[['Country/Region','Confirmed','WHO Region',]].head(50)
top_confirmed_cases
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x = top_confirmed_cases['Country/Region'],y = top_confirmed_cases['Confirmed'],data = top_confirmed_cases,palette = 'Oranges_r' )
plt.xlabel('Country/Region')
plt.ylabel('count in millions')
plt.title('Top confirmed cases')
# by looking at map US region is wide and almost first 25 country name include US so we will look at bottom confirmed cases
```

2.Let us have a look at bottom 10 covid confirmed country with there WHO region

```python
bottom_confirmed_cases = covid_19.sort_values('Confirmed',ascending = False )

bottom_confirmed_cases = covid_19 [['Country/Region','Confirmed','WHO Region']].tail(20)

bottom_confirmed_cases
```

```python
# plotting for bottom_confirmed_cases

plt.figure(figsize=(15,10))
sns.barplot(x = bottom_confirmed_cases['Country/Region'],y = bottom_confirmed_cases['Confirmed'],data = bottom_confirmed_cases,palette = 'rocket' )
plt.xlabel('Country/Region')
plt.ylabel('count in millions')
plt.title('Bottom confirmed cases')
```

3.Let us have a look at top 10 covid deaths country wise with there WHO region

```python
top_deaths_cases = covid_19.sort_values('Deaths',ascending = False )
top_deaths_cases = top_deaths_cases[['Deaths','Country/Region','WHO Region']].head(20)
top_deaths_cases
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x= top_deaths_cases['Country/Region'],y= top_deaths_cases['Deaths'],data = top_deaths_cases,palette = 'rocket')
plt.xlabel('Country/Region')
plt.ylabel('count')
plt.title('Top Confirmed Deaths Cases')
```

Confirmed cases and confirmed deaths are in US Region more than any other region
