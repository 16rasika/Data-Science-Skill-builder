**Importing Libraries for plotting and graphing**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Vehical TypesÂ¶**

```python
type_counts=ele_vehicle_df['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(8,6))
type_counts.plot(kind='pie')
plt.title('Electric Vehicle Types')
plt.ylabel('')
plt.show()
```

**Most electric vehicle used**

```python
plt.figure(figsize=(8, 6))
ele_vehicle_df.Model.value_counts().head(5).plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.xlabel("Most electrical vehicles used");
```

**Electric Vehicle Population by Country**

```python
country_counts=ele_vehicle_df['County'].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
country_counts.plot(kind='bar')
plt.title('Electric Vehicle Population by Country')
plt.xlabel('County')
plt.ylabel('Count')
plt.show()
```

**Vehicle Population by State**

```python
state_counts = ele_vehicle_df['State'].value_counts().nlargest(10)
plt.figure(figsize=(8, 6))
state_counts.plot(kind='bar')
plt.title('Electric Vehicle Population by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

**Most population of elctrical vehicles**

```python
plt.figure(figsize=(8, 6))
ele_vehicle_df.City.value_counts().head(5).plot(kind='pie'))
plt.ylabel("")
plt.xlabel("Cities with most electrical vehicles");
```

Top Makers of Electric VehicleÂ¶

```python
make_counts=ele_vehicle_df['Make'].value_counts()
plt.figure(figsize=(8, 6))
ele_vehicle_df.Make.value_counts().head(5).plot(kind='pie')
plt.ylabel("")
plt.xlabel("Top Makes of Electric Vehicles");
```

**Model Year vs Quantity**

```python
plt.figure(figsize=(8, 6))
ele_vehicle_df.groupby("Model Year").size().plot(kind="bar")
plt.ylabel("Quantity");
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ele_vehicle_df['Model Year'], bins=10)
plt.title('Distribution of Electric Vehicle Model Years')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.show()
```
