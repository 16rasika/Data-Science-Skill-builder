```python
#checking for duplicates
final_ratings.drop_duplicates()
```

```python
final_pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
final_pt
```

610 books and 672 users

```python
#making format readable und usable
final_pt.fillna(0,inplace=True)
```

```python
final_pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#similarity_scores = cosine_similarity(pt)
#similarity_scores.shape
```

```python
similarity_scores = cosine_similarity(final_pt)
similarity_scores
```

```python
## Function to recommend books
def recommend(book_name):
     # index fetch
     index = np.where(final_pt.index == book_name)[0][0]
     #distances = similarity_scores[index]
     similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True)[1:6]

     for i in similar_items:
         #print(i[0])  # here i am getting indexes of book
          print(final_pt.index[i[0]])

     #return suggestion
```

**Testing function**

```python
recommend('Message in a Bottle')
```
