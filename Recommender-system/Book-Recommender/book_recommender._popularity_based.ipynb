 **Popularity Based Recommender**

```python
ratings_books=ratings.merge(books,on='ISBN')
```

```python
ratings_books
```

```python
ratings_books.groupby('Book-Title').count()
```

```python
ratings_books.groupby('Book-Title').count()['Book-Rating']
```

```python
num_rating_df=ratings_books.groupby('Book-Title').count().reset_index() # with rest_index getting df
num_rating_df.rename(columns={'Book-Rating':'NumRating'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_books.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avgRating'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df[popular_df['NumRating']>=250]
```

```python
popular_df[popular_df['NumRating']>=250].sort_values('avgRating',ascending= False)
```

```python
#Top 50 popular book
popular_df[popular_df['NumRating']>=250].sort_values('avgRating',ascending= False).head(50)
```
