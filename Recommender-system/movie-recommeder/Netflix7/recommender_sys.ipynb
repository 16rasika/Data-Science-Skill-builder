Feature Selection
keeping only show_id,title,listed_in is actually genre and description 

```python
movies = movies[['show_id','title','listed_in','description']]
```

```python
movies
```

```python
np.warnings.filterwarnings('ignore')   # for avoiding warnings
```

```python
#quote_df.rename(columns={'A':'STK', 'B':'TOpen', 'C':'TPCLOSE', 'D':'TPrice', 'E':'THigh', 'F':'TLow', 'I':'TVol', 'J':'TAmt', 'e':'TDate', 'f':'TTime'}, inplace=True)
#quote_df = quote_df.ix[:,[0,3,2,1,4,5,8,9,30,31]]

movies.rename(columns={'show_id':'Id','title':'Title','listed_in':'Genre','description':'Overview'},inplace=True)
```

```python
movies
```

```python
movies['tags']= movies['Genre'] + movies['Overview']
```

```python
movies
```

As i combined Overview and Genre in one column Tags,i am going to drop the columns

```python
new_movies = movies.drop(columns=['Overview','Genre'])
new_movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()   # convering tags into array
```

```python
vector= cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity =cosine_similarity(vector)
```

```python
def recommand(movies):
    index=new_movies[new_movies['Title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_movies.iloc[i[0]].Title)
```

```python
recommand('Zombie Dumb')
```
